```python
import pandas as pd
import numpy as np
```

```python
df_mortality = pd.read_csv("../data_raw/child_mortality.csv")
df_mortality
```

```python
df_mortality = df_mortality.drop('Country Code', axis=1)
```

```python
df_mortality = df_mortality.drop('Indicator Code', axis=1)
```

```python
df_mortality.columns = df_mortality.columns.str.lower().str.replace(' ', '_')
df_mortality
```

```python
df_mortality = df_mortality.drop('indicator_name', axis=1)
df_mortality
```

```python
# Step 1: Melt the DataFrame to long format
df_melted = df_mortality.melt(id_vars='country_name', var_name='year', value_name='child_mortality')

# Step 2: Pivot the table so that countries become columns and years are rows
df_mortality_transposed = df_melted.pivot(index='year', columns='country_name', values='child_mortality')

# Optional: reset the index if needed
df_mortality_transposed.reset_index(inplace=True)
```

```python
df_mortality_transposed
```

```python
df_mortality_transposed = df_mortality_transposed.drop(index=64)
df_mortality_transposed
```

```python
#REMOVE NULL VALUES

# 1. Interpolate based on years
df_mortality_transposed.set_index('year', inplace=True)
df_mortality_transposed = df_mortality_transposed.interpolate(method='linear', axis=0)

# 2. Fill any remaining nulls with forward/backward fill
ddf_mortality_transposed = df_mortality_transposed.ffill().bfill()

# 3. Reset index if needed
df_mortality_transposed.reset_index(inplace=True)
```

```python
#Check if any null values are left

df_mortality_transposed.isnull().sum().sum()
```

```python
#Check which column consist null values
null_counts = df_mortality_transposed.isnull().sum()
null_columns = null_counts[null_counts > 0]
print(null_columns)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))
plt.plot(df_mortality_transposed['year'], df_mortality_transposed['World'], marker='o')
plt.xlabel('Year')
plt.ylabel('World')
plt.title('Whole Child Mortality Over Time')
plt.grid(True)
plt.show()
```

```python
#Top population countries check!!!

latest_year = df_mortality_transposed['year'].max()
latest_data = df_mortality_transposed[df_mortality_transposed['year'] == latest_year].drop(columns='year')
top5 = latest_data.T.sort_values(by=latest_data.index[0], ascending=False).head(40)
top5
```

```python
#Population visualization per specific countries
countries = ['China', 'India', 'United States', 'Nigeria', 'Indonesia']

plt.figure(figsize=(12, 6))
for country in countries:
    plt.plot(df_mortality_transposed['year'], df_mortality_transposed[country], label=country)

plt.title('Population Trends (1960–2023)')
plt.xlabel('Year')
plt.ylabel('Population')
plt.xticks(rotation=90)
plt.legend()
plt.grid(True)
plt.show()
```
