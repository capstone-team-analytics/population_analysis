```python
import pandas as pd
import numpy as np
```

```python
df_education = pd.read_csv("../data_raw/education.csv")
df_education
```

```python
df_education = df_education.drop('Country Code', axis=1)
```

```python
df_education = df_education.drop('Indicator Code', axis=1)
```

```python
df_education = df_education.drop('Indicator Name', axis=1)
```

```python
df_education.columns = df_education.columns.str.lower().str.replace(' ', '_')
df_education
```

```python
# Step 1: Melt the DataFrame to long format
df_melted_education = df_education.melt(id_vars='country_name', var_name='year', value_name='fertility')

# Step 2: Pivot the *melted* DataFrame
df_education_transposed = df_melted_education.pivot(index='year', columns='country_name', values='fertility')

# Optional: reset the index if needed
df_education_transposed.reset_index(inplace=True)
```

```python
df_education_transposed
```

```python
df_education_transposed = df_education_transposed.drop(index=64)
df_education_transposed
```

```python
#Remove null values
# Set 'year' as index
df_education_transposed.set_index('year', inplace=True)

# Interpolate missing values linearly across years
df_education_transposed = df_education_transposed.interpolate(method='linear', axis=0)

# Fill any remaining nulls
df_education_transposed = df_education_transposed.ffill().bfill()

# Reset index
df_education_transposed.reset_index(inplace=True)
```

```python
#Check if any null values are left

df_education_transposed.isnull().sum().sum()
```

```python
#Check which column consist null values
null_counts = df_education_transposed.isnull().sum()
null_columns = null_counts[null_counts > 0]
print(null_columns)
```

```python
df_education_transposed.info()
```

```python
df_education_transposed['year'] = df_education_transposed['year'].astype(int)
```

```python
#Top fertility countries check!!!

latest_year = df_education_transposed['year'].max()
latest_data = df_education_transposed[df_education_transposed['year'] == latest_year].drop(columns='year')
top5 = latest_data.T.sort_values(by=latest_data.index[0], ascending=False).head(40)
top5
```

```python
# 1. Set 'year' as the index
df_education = df_education_transposed.set_index('year')

# 2. Transpose so rows become countries and columns become years
df_education = df_education.T

# 3. Reset the index so that 'country_name' becomes a column
df_education = df_education.reset_index()

# 4. Rename the newly created 'index' column to 'country_name'
df_education = df_education.rename(columns={'index': 'country_name'})

df_education
```
