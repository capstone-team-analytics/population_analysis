```python
import pandas as pd
import numpy as np
```

```python
df_population = pd.read_csv("../data/population.csv")
df_population
```

```python
df_population = df_population.drop('Country Code', axis=1)
```

```python
df_population = df_population.drop('Indicator Code', axis=1)
```

```python
df_population.columns = df_population.columns.str.lower().str.replace(' ', '_')
df_population
```

```python
df_population = df_population.drop('indicator_name', axis=1)
df_population
```

```python
# Step 1: Melt the DataFrame to long format
df_melted = df_population.melt(id_vars='country_name', var_name='year', value_name='population')

# Step 2: Pivot the table so that countries become columns and years are rows
df_population_transposed = df_melted.pivot(index='year', columns='country_name', values='population')

# Optional: reset the index if needed
df_population_transposed.reset_index(inplace=True)
```

```python
df_population_transposed
```

```python
df_population_transposed = df_population_transposed.drop(index=64)
df_population_transposed
```

```python
df_population_transposed.to_csv('final_population_data.csv', index=False)
```

```python
#REMOVE NULL VALUES

# 1. Interpolate based on years
df_population_transposed.set_index('year', inplace=True)
df_population_transposed = df_population_transposed.interpolate(method='linear', axis=0)

# 2. Fill any remaining nulls with forward/backward fill
df_population_transposed = df_population_transposed.ffill().bfill()

# 3. Reset index if needed
df_population_transposed.reset_index(inplace=True)
```

```python
#Check if any null values are left

df_population_transposed.isnull().sum().sum()
```

```python
#Check which column consist null values
null_counts = df_population_transposed.isnull().sum()
null_columns = null_counts[null_counts > 0]
print(null_columns)
```

```python
#Remove the column filled with null values
df_population_transposed = df_population_transposed.drop(columns=['Not classified'])
```

```python
#Check if there any null values left
df_population_transposed.isnull().sum().sum()
```

```python
df_population_transposed.info()
```

```python
df_population_transposed['year'] = df_population_transposed['year'].astype(int)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))
plt.plot(df_population_transposed['year'], df_population_transposed['World'], marker='o')
plt.xlabel('Year')
plt.ylabel('World')
plt.title('Whole Population Over Time')
plt.grid(True)
plt.show()
```

```python
#Top population countries check!!!

latest_year = df_population_transposed['year'].max()
latest_data = df_population_transposed[df_population_transposed['year'] == latest_year].drop(columns='year')
top5 = latest_data.T.sort_values(by=latest_data.index[0], ascending=False).head(40)
top5
```

```python
#Population visualization per specific countries
countries = ['China', 'India', 'United States', 'Nigeria', 'Indonesia']

plt.figure(figsize=(12, 6))
for country in countries:
    plt.plot(df_population_transposed['year'], df_population_transposed[country], label=country)

plt.title('Population Trends (1960–2023)')
plt.xlabel('Year')
plt.ylabel('Population')
plt.xticks(rotation=90)
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 1. Set 'year' as the index
df_population = df_population_transposed.set_index('year')

# 2. Transpose so rows become countries and columns become years
df_population = df_population.T

# 3. Reset the index so that 'country_name' becomes a column
df_population = df_population.reset_index()

# 4. Rename the newly created 'index' column to 'country_name'
df_population = df_population.rename(columns={'index': 'country_name'})

df_population
```
