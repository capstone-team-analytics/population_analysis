```python
import pandas as pd
import numpy as np
```

```python
df_gdp = pd.read_csv("../data/gdp.csv")
df_gdp
```

```python
df_gdp = df_gdp.drop('Country Code', axis=1)
```

```python
df_gdp = df_gdp.drop('Indicator Code', axis=1)
```

```python
df_gdp = df_gdp.drop('Indicator Name', axis=1)
```

```python
df_gdp.columns = df_gdp.columns.str.lower().str.replace(' ', '_')
df_gdp
```

```python
# Step 1: Melt the DataFrame to long format
df_melted_gdp = df_gdp.melt(id_vars='country_name', var_name='year', value_name='gdp')

# Step 2: Pivot the *melted* DataFrame
df_gdp_transposed = df_melted_gdp.pivot(index='year', columns='country_name', values='gdp')

# Optional: reset the index if needed
df_gdp_transposed.reset_index(inplace=True)
```

```python
df_gdp_transposed
```

```python
df_gdp_transposed = df_gdp_transposed.drop(index=64)
df_gdp_transposed
```

```python
#Remove null values
# Set 'year' as index
df_gdp_transposed.set_index('year', inplace=True)

# Interpolate missing values linearly across years
df_gdp_transposed = df_gdp_transposed.interpolate(method='linear', axis=0)

# Fill any remaining nulls
df_gdp_transposed = df_gdp_transposed.ffill().bfill()

# Reset index
df_gdp_transposed.reset_index(inplace=True)
```

```python
#Check if any null values are left

df_gdp_transposed.isnull().sum().sum()
```

```python
#Check which column consist null values
null_counts = df_gdp_transposed.isnull().sum()
null_columns = null_counts[null_counts > 0]
print(null_columns)
```

```python
#Remove the column filled with null values
df_gdp_transposed = df_gdp_transposed.drop(columns=['Not classified'])
```

```python
#Check if there any null values left
df_gdp_transposed.isnull().sum().sum()
```

```python
df_gdp_transposed.info()
```

```python
df_gdp_transposed['year'] = df_gdp_transposed['year'].astype(int)
```

```python
print(df_gdp_transposed.describe())
```

```python
df_gdp_transposed.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))
plt.plot(df_gdp_transposed['year'], df_gdp_transposed['World'], marker='o')
plt.xlabel('Year')
plt.ylabel('World')
plt.title('Whole GDP Over Time')
plt.grid(True)
plt.show()
```

```python
#Top gdp countries check!!!

latest_year = df_gdp_transposed['year'].max()
latest_data = df_gdp_transposed[df_gdp_transposed['year'] == latest_year].drop(columns='year')
top5 = latest_data.T.sort_values(by=latest_data.index[0], ascending=False).head(40)
top5
```

```python
#Population visualization per specific countries
countries = ['China', 'India', 'United States', 'Nigeria', 'Indonesia']

plt.figure(figsize=(12, 6))
for country in countries:
    plt.plot(df_gdp_transposed['year'], df_gdp_transposed[country], label=country)

plt.title('GDP Trends (1960–2023)')
plt.xlabel('Year')
plt.ylabel('GDP rate')
plt.xticks(rotation=90)
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 1. Set 'year' as the index
df_gdp = df_gdp_transposed.set_index('year')

# 2. Transpose so rows become countries and columns become years
df_gdp = df_gdp.T

# 3. Reset the index so that 'country_name' becomes a column
df_gdp = df_gdp.reset_index()

# 4. Rename the newly created 'index' column to 'country_name'
df_gdp = df_gdp.rename(columns={'index': 'country_name'})

df_gdp
```
