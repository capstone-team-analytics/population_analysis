```python
import pandas as pd
import numpy as np
```

```python
df_fertility = pd.read_csv("../data_raw/fertility_rate.csv")
df_fertility
```

```python
df_fertility = df_fertility.drop('Country Code', axis=1)
```

```python
df_fertility = df_fertility.drop('Indicator Code', axis=1)
```

```python
df_fertility = df_fertility.drop('Indicator Name', axis=1)
```

```python
df_fertility.columns = df_fertility.columns.str.lower().str.replace(' ', '_')
df_fertility
```

```python
# Step 1: Melt the DataFrame to long format
df_melted_fert = df_fertility.melt(id_vars='country_name', var_name='year', value_name='fertility')

# Step 2: Pivot the *melted* DataFrame
df_fertility_transposed = df_melted_fert.pivot(index='year', columns='country_name', values='fertility')

# Optional: reset the index if needed
df_fertility_transposed.reset_index(inplace=True)
```

```python
df_fertility_transposed
```

```python
df_fertility_transposed = df_fertility_transposed.drop(index=64)
df_fertility_transposed
```

```python
#Remove null values
# Set 'year' as index
df_fertility_transposed.set_index('year', inplace=True)

# Interpolate missing values linearly across years
df_fertility_transposed = df_fertility_transposed.interpolate(method='linear', axis=0)

# Fill any remaining nulls
df_fertility_transposed = df_fertility_transposed.ffill().bfill()

# Reset index
df_fertility_transposed.reset_index(inplace=True)
```

```python
#Check if any null values are left

df_fertility_transposed.isnull().sum().sum()
```

```python
#Check which column consist null values
null_counts = df_fertility_transposed.isnull().sum()
null_columns = null_counts[null_counts > 0]
print(null_columns)
```

```python
#Remove the column filled with null values
df_population_transposed = df_fertility_transposed.drop(columns=['Not classified'])
```

```python
#Check if there any null values left
df_fertility_transposed.isnull().sum().sum()
```

```python
df_fertility_transposed.info()
```

```python
df_fertility_transposed['year'] = df_population_transposed['year'].astype(int)
```

```python
print(df_fertility_transposed.describe())
```

```python
df_fertility_transposed.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))
plt.plot(df_fertility_transposed['year'], df_fertility_transposed['World'], marker='o')
plt.xlabel('Year')
plt.ylabel('World')
plt.title('Whole Fertility Over Time')
plt.grid(True)
plt.show()
```

```python
#Top fertility countries check!!!

latest_year = df_fertility_transposed['year'].max()
latest_data = df_fertility_transposed[df_fertility_transposed['year'] == latest_year].drop(columns='year')
top5 = latest_data.T.sort_values(by=latest_data.index[0], ascending=False).head(40)
top5
```

```python
#Population visualization per specific countries
countries = ['China', 'India', 'United States', 'Nigeria', 'Indonesia']

plt.figure(figsize=(12, 6))
for country in countries:
    plt.plot(df_fertility_transposed['year'], df_fertility_transposed[country], label=country)

plt.title('Fertility Trends (1960–2023)')
plt.xlabel('Year')
plt.ylabel('Fertility rate')
plt.xticks(rotation=90)
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 1. Set 'year' as the index
df_fertility = df_fertility_transposed.set_index('year')

# 2. Transpose so rows become countries and columns become years
df_fertility = df_fertility.T

# 3. Reset the index so that 'country_name' becomes a column
df_fertility = df_fertility.reset_index()

# 4. Rename the newly created 'index' column to 'country_name'
df_fertility = df_fertility.rename(columns={'index': 'country_name'})

df_fertility
```
